```python
import pandas as pd

```

```python
df = pd.read_csv("hospital-costs.csv")
```

```python
df.head(2)
```

```python
df['Total Charges'] = df['Discharges'] * df['Mean Charge']
```

```python
df['Total Costs'] = df['Discharges'] * df['Mean Cost']
```

```python
df['Markup'] = df['Total Charges'] / df['Total Costs']
```
